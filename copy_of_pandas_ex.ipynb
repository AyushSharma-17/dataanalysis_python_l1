<h2>Dataframes</h2>

when we combine multiple series into a single frame called <b>dataframe</b>

```python
data={
    'name':['john','anna','peter','linda'],
    'age':[34,21,19,20],
    'city':['new york','paris','berlin','london'],
    'salary':[25000,32000,12500,18000]
}

df= pd.DataFrame(data)
```

```python
df
```
