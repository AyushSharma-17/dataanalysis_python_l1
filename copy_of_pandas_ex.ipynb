<h2>Removing the missing values</h2>

```python
df2.dropna()
```

```python
df2
```

```python
#removing the rows where non null values are at least =3
df2.dropna(thresh=3)
```

filling missing data

```python
df.fillna(0)
```

```python
df2.fillna(0)
```

```python
#fillling missing values with the specfic values
values={'A':0,'B':1,'C':2,'D':3}
df2.fillna(value=values)
```

```python
#filling null values with mean values
df2.fillna(df2.mean())
```

<h2>Merging joining concatination</h2>

```python
employees=pd.DataFrame({
    'employee_id':[1,2,3,4],
    'employee_name':['john','anna','peter','linda'],
    'department':['IT','HR','Finance','Marketing']
})

#dataFrame 2

salaries=pd.DataFrame({
    'employee_id':[1,2,3,6],
    'salary':[25000,32000,12500,18000],
    'bonus':[5000,7000,2000,3000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id')
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

<h2>Concatination of 2 dataframe</h2>

```python
df3=pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2']
})

df4=pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
})
```

```python
df3
```

```python
df4
```

```python
pd.concat([df3,df4])
```

```python
pd.concat([df3,df4] ,axis=1)
```

<h2>Joining 2 dataframes</h2>

```python
df5=pd.DataFrame({
    'name':['john','anna','peter','linda'],
    'index':[1,2,3,4]
})
#second dataframe
df6= pd.DataFrame({
    'score':[85,90,78,92],
    'index':[2,3,4,5]
})
```

```python
df5
```

```python
df6
```

```python
df5.merge(df6, on="index", how="inner")

```
